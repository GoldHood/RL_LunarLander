```python
# --- Imports y config básica
import math, random, time
from collections import deque, namedtuple
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
import gymnasium as gym
from tqdm import trange

# Reutilizamos el ENV_ID y env que ya creaste con pick_lander_env()
try:
    ENV_ID
except NameError:
    # fallback por si abres un cuaderno nuevo
    def pick_lander_env():
        for env_id in ("LunarLander-v3", "LunarLander-v2"):
            try:
                return env_id, gym.make(env_id)
            except Exception:
                continue
        gym.pprint_registry(pattern=r"LunarLander.*")
        raise RuntimeError("No se encontró LunarLander.")
    ENV_ID, env = pick_lander_env()
print("Usando:", ENV_ID)

# Seeds y device
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# Dimensiones
STATE_DIM = env.observation_space.shape[0]   # 8
ACT_DIM   = env.action_space.n               # 4

# Límites “seguros” para clipear observaciones
LOW  = np.array([-1.5, -0.5, -2.0, -2.0, -math.pi, -3.0, 0.0, 0.0], dtype=np.float32)
HIGH = np.array([ 1.5,  1.5,  2.0,  2.0,  math.pi,  3.0, 1.0, 1.0], dtype=np.float32)

def clip_obs(s: np.ndarray) -> np.ndarray:
    s = np.asarray(s, dtype=np.float32)
    s[:6] = np.clip(s[:6], LOW[:6], HIGH[:6])    # patas ya están en [0,1]
    s[6:] = np.clip(s[6:], 0.0, 1.0)
    return s

Transition = namedtuple("Transition", ["s","a","r","sp","d"])

class ReplayBuffer:
    def __init__(self, capacity=300_000):
        self.buf = deque(maxlen=capacity)
    def __len__(self): return len(self.buf)
    def push(self, *args): self.buf.append(Transition(*args))
    def sample(self, batch_size, device=device):
        batch = random.sample(self.buf, batch_size)
        s  = torch.tensor(np.stack([b.s for b in batch]), dtype=torch.float32, device=device)
        a  = torch.tensor([b.a for b in batch], dtype=torch.long, device=device)
        r  = torch.tensor([b.r for b in batch], dtype=torch.float32, device=device)
        sp = torch.tensor(np.stack([b.sp for b in batch]), dtype=torch.float32, device=device)
        d  = torch.tensor([b.d for b in batch], dtype=torch.float32, device=device)
        return s, a, r, sp, d

# Buffers (real y sintético)
real_buffer  = ReplayBuffer(300_000)
model_buffer = ReplayBuffer(300_000)
```

```python
class QNet(nn.Module):
    def __init__(self, s_dim=STATE_DIM, a_dim=ACT_DIM, hidden=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(s_dim, hidden), nn.ReLU(),
            nn.Linear(hidden, hidden), nn.ReLU(),
            nn.Linear(hidden, a_dim),
        )
    def forward(self, s):  # s: [B, s_dim]
        return self.net(s)

class DQNAgent:
    def __init__(self, gamma=0.99, lr=3e-4, eps_start=1.0, eps_end=0.05, eps_decay=2e-5, tau=0.005):
        self.q = QNet().to(device)
        self.target = QNet().to(device)
        self.target.load_state_dict(self.q.state_dict())
        self.opt = torch.optim.Adam(self.q.parameters(), lr=lr)
        self.gamma = gamma
        self.eps = eps_start; self.eps_end = eps_end; self.eps_decay = eps_decay
        self.tau = tau
        self.step_count = 0

    def act(self, s_np):
        self.step_count += 1
        self.eps = max(self.eps_end, self.eps - self.eps_decay)
        if random.random() < self.eps:
            return random.randrange(ACT_DIM)
        s = torch.tensor(s_np, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            return int(self.q(s).argmax(1).item())

    def train_step(self, batch, clip_grad=1.0):
        s, a, r, sp, d = batch
        qsa = self.q(s).gather(1, a.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            max_next = self.target(sp).max(1).values
            target = r + self.gamma * (1.0 - d) * max_next
        loss = F.smooth_l1_loss(qsa, target)
        self.opt.zero_grad(); loss.backward()
        torch.nn.utils.clip_grad_norm_(self.q.parameters(), clip_grad)
        self.opt.step()
        # soft update
        with torch.no_grad():
            for p, tp in zip(self.q.parameters(), self.target.parameters()):
                tp.data.mul_(1 - self.tau).add_(self.tau * p.data)
        return float(loss.item())
```

```python
class DynamicsNet(nn.Module):
    def __init__(self, s_dim=STATE_DIM, a_dim=ACT_DIM, hidden=256):
        super().__init__()
        self.fc1 = nn.Linear(s_dim + a_dim, hidden)
        self.fc2 = nn.Linear(hidden, hidden)
        self.head_ds = nn.Linear(hidden, s_dim)  # delta s
        self.head_r  = nn.Linear(hidden, 1)     # reward
        self.head_d  = nn.Linear(hidden, 1)     # done logit

    def forward(self, s, a_onehot):  # s: [B,8], a_onehot: [B,4]
        x = torch.cat([s, a_onehot], dim=-1)
        x = F.relu(self.fc1(x)); x = F.relu(self.fc2(x))
        ds = self.head_ds(x)
        r  = self.head_r(x).squeeze(-1)
        d  = torch.sigmoid(self.head_d(x)).squeeze(-1)
        return ds, r, d

class DynamicsEnsemble:
    def __init__(self, E=5, s_dim=STATE_DIM, a_dim=ACT_DIM, lr=3e-4):
        self.models = [DynamicsNet(s_dim, a_dim).to(device) for _ in range(E)]
        self.optims = [torch.optim.Adam(m.parameters(), lr=lr) for m in self.models]
        self.a_dim = a_dim

    def train_steps(self, buffer: ReplayBuffer, steps=800, batch_size=512):
        if len(buffer) < batch_size: return 0.0
        losses = []
        for _ in range(steps):
            s, a, r, sp, d = buffer.sample(batch_size)
            a_oh = F.one_hot(a, num_classes=self.a_dim).float()
            tgt_ds = sp - s
            for m, opt in zip(self.models, self.optims):
                pred_ds, pred_r, pred_d = m(s, a_oh)
                loss = F.mse_loss(pred_ds, tgt_ds) \
                     + F.mse_loss(pred_r, r) \
                     + F.binary_cross_entropy(pred_d, d)
                opt.zero_grad(); loss.backward(); opt.step()
                losses.append(loss.item())
        return float(np.mean(losses)) if losses else 0.0

    @torch.no_grad()
    def predict(self, s, a):
        a_oh = F.one_hot(a, num_classes=self.a_dim).float()
        sp_list, r_list, d_list = [], [], []
        for m in self.models:
            ds, r, d = m(s, a_oh)
            sp_list.append(s + ds); r_list.append(r); d_list.append(d)
        sp = torch.stack(sp_list, 0)   # [E,B,8]
        rr = torch.stack(r_list,  0)   # [E,B]
        dd = torch.stack(d_list,  0)   # [E,B]
        sp_m, rr_m, dd_m = sp.mean(0), rr.mean(0), dd.mean(0)
        # incertidumbre: norma del std de s' + std de r
        uncert = sp.std(0).norm(dim=1) + rr.std(0)
        return sp_m, rr_m, dd_m, uncert
```

```python
def rollout_model(agent: DQNAgent, world: DynamicsEnsemble,
                  start_states: torch.Tensor,  # [B,8] en device
                  horizon=2, beta_uncert=1.0, eps_model=0.05,
                  max_per_state=2):
    B = start_states.size(0)
    s = start_states.clone()
    out = []
    for _ in range(horizon):
        with torch.no_grad():
            q = agent.q(s)
            a = q.argmax(1)
            mask = torch.rand(B, device=device) < eps_model
            if mask.any():
                a[mask] = torch.randint(0, ACT_DIM, (int(mask.sum()),), device=device)
        sp, r, d, u = world.predict(s, a)
        r_pen = r - beta_uncert * u
        s_np, sp_np, r_np, d_np = s.cpu().numpy(), sp.cpu().numpy(), r_pen.cpu().numpy(), d.cpu().numpy()
        for i in range(B):
            tr = Transition(
                s=clip_obs(s_np[i]),
                a=int(a[i].item()),
                r=float(r_np[i]),
                sp=clip_obs(sp_np[i]),
                d=float(d_np[i] > 0.5),
            )
            out.append(tr)
        s = sp
    return out[:B*max_per_state]
```

```python
def train_worldmodels_lander(
    env_id=ENV_ID,
    episodes=600,
    warmup_steps=5_000,
    train_every=1,
    batch_size=256,
    real_batch_frac=0.6,
    wm_train_every=2_000,
    wm_train_steps=800,
    rollout_every=2_000,
    rollout_horizon=2,
    rollout_start_states=1024,
):
    agent = DQNAgent()
    world = DynamicsEnsemble(E=5)
    real_buf  = ReplayBuffer(300_000)
    model_buf = ReplayBuffer(300_000)

    env = gym.make(env_id)
    rewards = []
    total_steps = 0
    state, _ = env.reset(seed=SEED)
    state = clip_obs(state)

    pbar = trange(episodes, desc="Training", ncols=120)
    for ep in pbar:
        done = False; ep_ret = 0.0
        while not done:
            a = agent.act(state)
            next_state, r, terminated, truncated, _ = env.step(a)
            d = terminated or truncated
            next_state = clip_obs(next_state)

            real_buf.push(state, a, r, next_state, float(d))
            state = next_state; ep_ret += r; total_steps += 1; done = d

            # Entrenamiento DQN con mezcla real+modelo
            if total_steps > warmup_steps and total_steps % train_every == 0:
                rb = int(batch_size * real_batch_frac)
                mb = batch_size - rb
                if len(real_buf) >= rb and len(model_buf) >= mb:
                    br = real_buf.sample(rb)
                    bm = model_buf.sample(mb)
                    batch = (
                        torch.cat([br[0], bm[0]], 0),
                        torch.cat([br[1], bm[1]], 0),
                        torch.cat([br[2], bm[2]], 0),
                        torch.cat([br[3], bm[3]], 0),
                        torch.cat([br[4], bm[4]], 0),
                    )
                    loss = agent.train_step(batch)

            # Entrenar World Model
            if total_steps >= warmup_steps and total_steps % wm_train_every == 0:
                wm_loss = world.train_steps(real_buf, steps=wm_train_steps, batch_size=512)

            # Rollouts sintéticos cortos
            if total_steps >= warmup_steps and total_steps % rollout_every == 0 and len(real_buf) >= rollout_start_states:
                s0, _, _, _, _ = real_buf.sample(rollout_start_states)
                synth = rollout_model(agent, world, s0, horizon=rollout_horizon, beta_uncert=1.0, eps_model=0.05)
                for tr in synth:
                    model_buf.push(tr.s, tr.a, tr.r, tr.sp, tr.d)

        rewards.append(ep_ret)
        state, _ = env.reset()
        state = clip_obs(state)
        avg100 = np.mean(rewards[-100:])
        pbar.set_postfix(
            ep_ret=f"{ep_ret:7.1f}",
            avg100=f"{avg100:7.1f}",
            steps=total_steps,
            eps=f"{agent.eps:.3f}",
            real=len(real_buf),
            model=len(model_buf),
        )

    env.close()
    return rewards, agent, world

rewards, agent, world = train_worldmodels_lander(episodes=400)  # empieza con 400 para probar
```

```python
import matplotlib.pyplot as plt

def evaluate(agent: DQNAgent, env_id=ENV_ID, episodes=10):
    env_eval = gym.make(env_id)
    scores = []
    for _ in range(episodes):
        s, _ = env_eval.reset(seed=random.randint(0, 10_000))
        s = clip_obs(s)
        done = False; total = 0.0
        while not done:
            with torch.no_grad():
                q = agent.q(torch.tensor(s, dtype=torch.float32, device=device).unsqueeze(0))
                a = int(q.argmax(1).item())
            sp, r, term, trunc, _ = env_eval.step(a)
            total += r; s = clip_obs(sp); done = term or trunc
        scores.append(total)
    env_eval.close()
    return float(np.mean(scores)), scores

def plot_rewards(rewards):
    plt.figure(figsize=(8,4))
    plt.plot(rewards, alpha=0.5, label="Return por episodio")
    if len(rewards) >= 100:
        mov = np.convolve(rewards, np.ones(100)/100, mode="valid")
        plt.plot(range(99, 99+len(mov)), mov, label="Media móvil (100)")
    plt.xlabel("Episodio"); plt.ylabel("Return")
    plt.title(f"LunarLander — DQN + World Model ({ENV_ID})")
    plt.legend(); plt.grid(True); plt.show()

plot_rewards(rewards)
mean_eval, scores = evaluate(agent, ENV_ID, episodes=10)
print(f"Return promedio (eval 10 eps): {mean_eval:.1f}")
```

```python
# ==== CONFIG / IMPORTS ========================================================
import math, random, time
from collections import deque, namedtuple
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
import gymnasium as gym
from tqdm import trange
import matplotlib.pyplot as plt

SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ==== ENV PICKER (v3 si existe, si no v2) ====================================
def pick_lander_env():
    for env_id in ("LunarLander-v3", "LunarLander-v2"):
        try:
            env = gym.make(env_id)
            return env_id, env
        except Exception:
            continue
    gym.pprint_registry(pattern=r"LunarLander.*")
    raise RuntimeError("No se encontró un LunarLander válido.")

ENV_ID, env = pick_lander_env()
print("Usando:", ENV_ID)

STATE_DIM = env.observation_space.shape[0]   # 8
ACT_DIM   = env.action_space.n               # 4

# ==== UTILS ===================================================================
LOW  = np.array([-1.5, -0.5, -2.0, -2.0, -math.pi, -3.0, 0.0, 0.0], dtype=np.float32)
HIGH = np.array([ 1.5,  1.5,  2.0,  2.0,  math.pi,  3.0, 1.0, 1.0], dtype=np.float32)

def clip_obs(s: np.ndarray) -> np.ndarray:
    s = np.asarray(s, dtype=np.float32)
    s[:6] = np.clip(s[:6], LOW[:6], HIGH[:6])  # patas ya están en [0,1]
    s[6:] = np.clip(s[6:], 0.0, 1.0)
    return s

Transition = namedtuple("Transition", ["s","a","r","sp","d"])

class ReplayBuffer:
    def __init__(self, capacity=400_000):
        self.buf = deque(maxlen=capacity)
    def __len__(self): return len(self.buf)
    def push(self, *args): self.buf.append(Transition(*args))
    def sample(self, batch_size, device=device):
        batch = random.sample(self.buf, batch_size)
        s  = torch.tensor(np.stack([b.s  for b in batch]), dtype=torch.float32, device=device)
        a  = torch.tensor([b.a for b in batch], dtype=torch.long, device=device)
        r  = torch.tensor([b.r for b in batch], dtype=torch.float32, device=device)
        sp = torch.tensor(np.stack([b.sp for b in batch]), dtype=torch.float32, device=device)
        d  = torch.tensor([b.d for b in batch], dtype=torch.float32, device=device)
        return s, a, r, sp, d

# ==== BASELINE DQN (Double DQN) ===============================================
class QNet(nn.Module):
    def __init__(self, s_dim=STATE_DIM, a_dim=ACT_DIM, hidden=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(s_dim, hidden), nn.ReLU(),
            nn.Linear(hidden, hidden), nn.ReLU(),
            nn.Linear(hidden, a_dim),
        )
    def forward(self, s): return self.net(s)

class DQNAgent:
    def __init__(self, gamma=0.99, lr=3e-4, eps_start=1.0, eps_end=0.05, eps_decay=2e-5, tau=0.005):
        self.q = QNet().to(device)
        self.target = QNet().to(device); self.target.load_state_dict(self.q.state_dict())
        self.opt = torch.optim.Adam(self.q.parameters(), lr=lr)
        self.gamma = gamma; self.eps = eps_start; self.eps_end = eps_end; self.eps_decay = eps_decay
        self.tau = tau; self.step_count=0

    def act(self, s_np):
        self.step_count += 1
        self.eps = max(self.eps_end, self.eps - self.eps_decay)
        if random.random() < self.eps:
            return random.randrange(ACT_DIM)
        s = torch.tensor(s_np, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad(): return int(self.q(s).argmax(1).item())

    # Double DQN backup
    def train_step(self, batch, clip_grad=1.0):
        s, a, r, sp, d = batch
        qsa = self.q(s).gather(1, a.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            next_a = self.q(sp).argmax(1, keepdim=True)
            next_q = self.target(sp).gather(1, next_a).squeeze(1)
            target = r + self.gamma * (1.0 - d) * next_q
        loss = F.smooth_l1_loss(qsa, target)
        self.opt.zero_grad(); loss.backward()
        torch.nn.utils.clip_grad_norm_(self.q.parameters(), clip_grad)
        self.opt.step()
        # soft update
        with torch.no_grad():
            for p, tp in zip(self.q.parameters(), self.target.parameters()):
                tp.data.mul_(1 - self.tau).add_(self.tau * p.data)
        return float(loss.item())
```

```python
# ==== WORLD MODEL: normalización + ensemble ===================================
class Scaler:
    def __init__(self): self.fitted=False
    def fit_from_buffer(self, buf: ReplayBuffer, sample=100_000):
        n = min(len(buf), sample)
        if n < 1000: return
        s, a, r, sp, d = buf.sample(n)
        s = s.cpu().numpy(); sp = sp.cpu().numpy(); r = r.cpu().numpy()
        ds = sp - s; eps = 1e-6
        self.mu_s,  self.std_s  = s.mean(0), s.std(0)+eps
        self.mu_ds, self.std_ds = ds.mean(0), ds.std(0)+eps
        self.mu_r,  self.std_r  = float(r.mean()), float(r.std()+eps)
        self.fitted=True

class DynamicsNet(nn.Module):
    def __init__(self, s_dim=STATE_DIM, a_dim=ACT_DIM, hidden=512):
        super().__init__()
        self.fc1 = nn.Linear(s_dim + a_dim, hidden)
        self.fc2 = nn.Linear(hidden, hidden)
        self.head_ds = nn.Linear(hidden, s_dim)
        self.head_r  = nn.Linear(hidden, 1)
        self.head_d  = nn.Linear(hidden, 1)
    def forward(self, s_n, a_oh):
        x = torch.cat([s_n, a_oh], -1)
        x = F.relu(self.fc1(x)); x = F.relu(self.fc2(x))
        ds_n = self.head_ds(x)
        r_n  = self.head_r(x).squeeze(-1)
        d    = torch.sigmoid(self.head_d(x)).squeeze(-1)
        return ds_n, r_n, d

class DynamicsEnsemble:
    def __init__(self, E=5, s_dim=STATE_DIM, a_dim=ACT_DIM, lr=3e-4, hidden=512, wd=1e-5):
        self.models = [DynamicsNet(s_dim, a_dim, hidden).to(device) for _ in range(E)]
        self.optims = [torch.optim.Adam(m.parameters(), lr=lr, weight_decay=wd) for m in self.models]
        self.a_dim = a_dim
        self.scaler = Scaler()

    def train_steps(self, buffer: ReplayBuffer, steps=1200, batch_size=512):
        if len(buffer) < batch_size: return 0.0
        self.scaler.fit_from_buffer(buffer)
        if not self.scaler.fitted: return 0.0
        mu_s  = torch.tensor(self.scaler.mu_s,  device=device)
        std_s = torch.tensor(self.scaler.std_s, device=device)
        mu_ds = torch.tensor(self.scaler.mu_ds, device=device)
        std_ds= torch.tensor(self.scaler.std_ds,device=device)
        losses=[]
        for _ in range(steps):
            s, a, r, sp, d = buffer.sample(batch_size)
            a_oh = F.one_hot(a, num_classes=self.a_dim).float()
            ds = sp - s
            s_n  = (s  - mu_s)  / std_s
            ds_n = (ds - mu_ds) / std_ds
            r_n  = (r  - self.scaler.mu_r) / self.scaler.std_r
            for m,opt in zip(self.models, self.optims):
                p_ds, p_r, p_d = m(s_n, a_oh)
                loss = F.mse_loss(p_ds, ds_n) + F.mse_loss(p_r, r_n) + F.binary_cross_entropy(p_d, d)
                opt.zero_grad(); loss.backward()
                torch.nn.utils.clip_grad_norm_(m.parameters(), 1.0)
                opt.step()
                losses.append(loss.item())
        return float(np.mean(losses)) if losses else 0.0

    @torch.no_grad()
    def predict(self, s, a):
        assert self.scaler.fitted, "Scaler no calibrado"
        mu_s  = torch.tensor(self.scaler.mu_s,  device=device)
        std_s = torch.tensor(self.scaler.std_s, device=device)
        mu_ds = torch.tensor(self.scaler.mu_ds, device=device)
        std_ds= torch.tensor(self.scaler.std_ds,device=device)
        a_oh = F.one_hot(a, num_classes=self.a_dim).float()
        s_n = (s - mu_s) / std_s
        sp_list, r_list, d_list = [], [], []
        for m in self.models:
            ds_n, r_n, d = m(s_n, a_oh)
            ds = ds_n * std_ds + mu_ds
            r  = r_n * self.scaler.std_r + self.scaler.mu_r
            sp_list.append(s + ds); r_list.append(r); d_list.append(d)
        sp = torch.stack(sp_list,0); rr=torch.stack(r_list,0); dd=torch.stack(d_list,0)
        sp_m, rr_m, dd_m = sp.mean(0), rr.mean(0), dd.mean(0)
        uncert = sp.std(0).norm(dim=1) + rr.std(0)
        return sp_m, rr_m, dd_m, uncert
```

```python
# ==== ROLLOUTS con filtro por incertidumbre ===================================
def rollout_model(agent: DQNAgent, world: DynamicsEnsemble,
                  start_states: torch.Tensor, horizon=1,
                  eps_model=0.05, keep_quantile=0.6):
    """Devuelve lista de Transition sintéticas; filtra el 40% más incierto."""
    B = start_states.size(0)
    s = start_states.clone()
    out=[]
    for _ in range(horizon):
        with torch.no_grad():
            q = agent.q(s); a = q.argmax(1)
            m = torch.rand(B, device=device) < eps_model
            if m.any(): a[m] = torch.randint(0, ACT_DIM, (int(m.sum()),), device=device)
        sp, r, d, u = world.predict(s, a)
        thr = torch.quantile(u, keep_quantile)  # conservar más confiables
        keep = u <= thr
        s_k, a_k, sp_k, r_k, d_k = s[keep], a[keep], sp[keep], r[keep], d[keep]
        s_np, sp_np = s_k.cpu().numpy(), sp_k.cpu().numpy()
        r_np, d_np = r_k.cpu().numpy(), d_k.cpu().numpy()
        for i in range(s_np.shape[0]):
            out.append(Transition(clip_obs(s_np[i]), int(a_k[i].item()),
                                  float(r_np[i]), clip_obs(sp_np[i]), float(d_np[i] > 0.5)))
        s = sp
    return out
```

```python
# ==== TRAIN LOOP (baseline o world model) =====================================
def plot_rewards(rewards, title=""):
    plt.figure(figsize=(8,4))
    plt.plot(rewards, alpha=0.5, label="Return por episodio")
    if len(rewards) >= 100:
        mov = np.convolve(rewards, np.ones(100)/100, mode="valid")
        plt.plot(range(99,99+len(mov)), mov, label="Media móvil (100)")
    plt.xlabel("Episodio"); plt.ylabel("Return"); plt.title(title); plt.grid(True); plt.legend(); plt.show()

def evaluate(agent: DQNAgent, env_id=ENV_ID, episodes=10):
    env = gym.make(env_id)
    scores=[]
    for _ in range(episodes):
        s,_ = env.reset(seed=random.randint(0,10_000)); s=clip_obs(s)
        done=False; total=0.0
        while not done:
            with torch.no_grad():
                a = int(agent.q(torch.tensor(s, dtype=torch.float32, device=device).unsqueeze(0)).argmax(1).item())
            sp, r, term, trunc, _ = env.step(a)
            total += r; s = clip_obs(sp); done = term or trunc
        scores.append(total)
    env.close()
    return float(np.mean(scores)), scores

def train_worldmodels_lander(
    env_id=ENV_ID,
    episodes=700,
    use_model=True,               # False = baseline DQN
    warmup_steps=10_000,          # pasos REALES antes de usar modelo
    batch_size=256,
    train_every=1,
    real_batch_frac=0.8,          # mezcla real:modelo en el batch
    wm_train_every=2_000,         # cada X pasos entrenamos world model
    wm_train_steps=1200,
    rollout_every=5_000,          # cada X pasos generamos rollouts
    rollout_start_states=2048,
    rollout_horizon=1,
):
    agent = DQNAgent()
    world = DynamicsEnsemble(E=5)
    real_buf  = ReplayBuffer(400_000)
    model_buf = ReplayBuffer(400_000)

    env = gym.make(env_id)
    rewards=[]
    total_steps=0
    s,_ = env.reset(seed=SEED); s=clip_obs(s)

    pbar = trange(episodes, desc=("DQN+WM" if use_model else "DQN baseline"), ncols=120)
    for ep in pbar:
        done=False; ep_ret=0.0
        while not done:
            a = agent.act(s)
            sp, r, term, trunc, _ = env.step(a)
            d = term or trunc
            sp = clip_obs(sp)
            real_buf.push(s,a,r,sp,float(d))
            s=sp; ep_ret+=r; total_steps+=1; done=d

            # Entrenar DQN (mezcla real + modelo si corresponde)
            if total_steps > warmup_steps and total_steps % train_every == 0:
                if not use_model:
                    if len(real_buf) >= batch_size:
                        agent.train_step(real_buf.sample(batch_size))
                else:
                    rb = int(batch_size * real_batch_frac)
                    mb = batch_size - rb
                    if len(real_buf) >= rb and len(model_buf) >= mb:
                        br = real_buf.sample(rb); bm = model_buf.sample(mb)
                        batch = (
                            torch.cat([br[0], bm[0]], 0),
                            torch.cat([br[1], bm[1]], 0),
                            torch.cat([br[2], bm[2]], 0),
                            torch.cat([br[3], bm[3]], 0),
                            torch.cat([br[4], bm[4]], 0),
                        )
                        agent.train_step(batch)

            # Entrenar world model con datos reales
            if use_model and total_steps >= warmup_steps and total_steps % wm_train_every == 0:
                _ = world.train_steps(real_buf, steps=wm_train_steps, batch_size=512)

            # Generar rollouts sintéticos cortos y filtrados
            if use_model and total_steps >= warmup_steps and total_steps % rollout_every == 0:
                if world.scaler.fitted and len(real_buf) >= rollout_start_states:
                    s0,_,_,_,_ = real_buf.sample(rollout_start_states)
                    synth = rollout_model(agent, world, s0, horizon=rollout_horizon,
                                          eps_model=0.05, keep_quantile=0.6)
                    for tr in synth:
                        model_buf.push(tr.s, tr.a, tr.r, tr.sp, tr.d)

        rewards.append(ep_ret)
        s,_ = env.reset(); s=clip_obs(s)
        avg100 = np.mean(rewards[-100:])
        pbar.set_postfix(avg100=f"{avg100:7.1f}", ep_ret=f"{ep_ret:7.1f}",
                         steps=total_steps, eps=f"{agent.eps:.3f}",
                         real=len(real_buf), model=len(model_buf))
    env.close()
    return rewards, agent, world

print("Código cargado ✓")
```

```python
# (A) Baseline DQN (sin modelo): referencia
rewards_base, agent_base, _ = train_worldmodels_lander(
    episodes=500,
    use_model=False,             # <-- sin world model
    warmup_steps=10_000,
)
plot_rewards(rewards_base, f"Baseline DQN — {ENV_ID}")
mean_eval_base, _ = evaluate(agent_base, ENV_ID, episodes=10)
print(f"[BASELINE] Return promedio (10 eps): {mean_eval_base:.1f}")
```

```python
# ============================ FAST TRAINER (VecEnv + Dueling DDQN + WM) ============================
import math, random, time
from collections import deque, namedtuple
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
import gymnasium as gym
from tqdm import trange
import matplotlib.pyplot as plt

# ----------------- Config general
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

def pick_lander_env():
    for env_id in ("LunarLander-v3","LunarLander-v2"):
        try:
            env = gym.make(env_id); env.close()
            return env_id
        except Exception:
            pass
    gym.pprint_registry(pattern=r"LunarLander.*")
    raise RuntimeError("No se encontró LunarLander.")
ENV_ID = pick_lander_env()
print("Usando:", ENV_ID)

# ----------------- Utilidades
LOW  = np.array([-1.5,-0.5,-2.0,-2.0,-math.pi,-3.0,0.0,0.0],dtype=np.float32)
HIGH = np.array([ 1.5, 1.5, 2.0, 2.0, math.pi, 3.0,1.0,1.0],dtype=np.float32)

def clip_obs(s: np.ndarray)->np.ndarray:
    s = np.asarray(s,dtype=np.float32)
    s[:6] = np.clip(s[:6],LOW[:6],HIGH[:6]); s[6:] = np.clip(s[6:],0.0,1.0)
    return s

Transition = namedtuple("Transition",["s","a","r","sp","d"])

class ReplayBuffer:
    def __init__(self, capacity=500_000): self.buf=deque(maxlen=capacity)
    def __len__(self): return len(self.buf)
    def push(self,*args): self.buf.append(Transition(*args))
    def sample(self,bs,device=device):
        batch=random.sample(self.buf,bs)
        s  = torch.tensor(np.stack([b.s for b in batch]),dtype=torch.float32,device=device)
        a  = torch.tensor([b.a for b in batch],dtype=torch.long,device=device)
        r  = torch.tensor([b.r for b in batch],dtype=torch.float32,device=device)
        sp = torch.tensor(np.stack([b.sp for b in batch]),dtype=torch.float32,device=device)
        d  = torch.tensor([b.d for b in batch],dtype=torch.float32,device=device)
        return s,a,r,sp,d

# ----------------- Dueling Double DQN
class DuelingQNet(nn.Module):
    def __init__(self,s_dim=8,a_dim=4,h=256):
        super().__init__()
        self.f = nn.Sequential(nn.Linear(s_dim,h), nn.ReLU(), nn.Linear(h,h), nn.ReLU())
        self.V = nn.Sequential(nn.Linear(h,h), nn.ReLU(), nn.Linear(h,1))
        self.A = nn.Sequential(nn.Linear(h,h), nn.ReLU(), nn.Linear(h,a_dim))
    def forward(self,s):
        x = self.f(s)
        v = self.V(x)
        a = self.A(x)
        return v + a - a.mean(dim=1,keepdim=True)

class DQNAgent:
    def __init__(self, s_dim=8, a_dim=4, gamma=0.99, lr=3e-4, eps_start=1.0, eps_end=0.05, eps_decay=2e-5, tau=0.01):
        self.q = DuelingQNet(s_dim,a_dim).to(device)
        self.target = DuelingQNet(s_dim,a_dim).to(device); self.target.load_state_dict(self.q.state_dict())
        self.opt = torch.optim.Adam(self.q.parameters(), lr=lr)
        self.gamma=gamma; self.eps=eps_start; self.eps_end=eps_end; self.eps_decay=eps_decay; self.tau=tau
    def act_batch(self, states_np):
        """Epsilon-greedy para un batch (vec env)."""
        B = states_np.shape[0]
        acts = np.empty(B, dtype=np.int64)
        self.eps = max(self.eps_end, self.eps - self.eps_decay*B)  # decay proporcional a transiciones
        with torch.no_grad():
            qs = self.q(torch.tensor(states_np,dtype=torch.float32,device=device))
            greedy = qs.argmax(1).detach().cpu().numpy()
        rnd = np.random.rand(B) < self.eps
        acts[:] = greedy
        acts[rnd] = np.random.randint(0, qs.shape[1], size=rnd.sum())
        return acts
    def train_step(self, batch, clip_grad=1.0):
        s,a,r,sp,d = batch
        qsa = self.q(s).gather(1,a.unsqueeze(1)).squeeze(1)
        with torch.no_grad():
            next_a = self.q(sp).argmax(1,keepdim=True)
            next_q = self.target(sp).gather(1,next_a).squeeze(1)
            tgt = r + self.gamma*(1.0-d)*next_q
        loss = F.smooth_l1_loss(qsa,tgt)
        self.opt.zero_grad(); loss.backward()
        torch.nn.utils.clip_grad_norm_(self.q.parameters(), clip_grad)
        self.opt.step()
        # soft target
        with torch.no_grad():
            for p,tp in zip(self.q.parameters(), self.target.parameters()):
                tp.data.mul_(1-self.tau).add_(self.tau*p.data)
        return float(loss.item())

# ----------------- World Model (ensemble + normalización)
class Scaler:
    def __init__(self): self.fitted=False
    def fit_from_buffer(self, buf: ReplayBuffer, sample=100_000):
        n=min(len(buf),sample)
        if n<2000: return
        s,a,r,sp,d = buf.sample(n)
        s,sp,r = s.cpu().numpy(), sp.cpu().numpy(), r.cpu().numpy()
        ds = sp - s; eps=1e-6
        self.mu_s, self.std_s   = s.mean(0),  s.std(0)+eps
        self.mu_ds,self.std_ds  = ds.mean(0), ds.std(0)+eps
        self.mu_r, self.std_r   = float(r.mean()), float(r.std()+eps)
        self.fitted=True

class DynNet(nn.Module):
    def __init__(self, s_dim=8, a_dim=4, h=512):
        super().__init__()
        self.fc1=nn.Linear(s_dim+a_dim,h); self.fc2=nn.Linear(h,h)
        self.ds=nn.Linear(h,s_dim); self.r=nn.Linear(h,1); self.d=nn.Linear(h,1)
    def forward(self,sn,aoh):
        x=torch.cat([sn,aoh],-1); x=F.relu(self.fc1(x)); x=F.relu(self.fc2(x))
        return self.ds(x), self.r(x).squeeze(-1), torch.sigmoid(self.d(x)).squeeze(-1)

class DynEnsemble:
    def __init__(self,E=5,s_dim=8,a_dim=4,lr=3e-4,wd=1e-5):
        self.models=[DynNet(s_dim,a_dim).to(device) for _ in range(E)]
        self.optims=[torch.optim.Adam(m.parameters(),lr=lr,weight_decay=wd) for m in self.models]
        self.a_dim=a_dim; self.scaler=Scaler()
    def train_steps(self, buf: ReplayBuffer, steps=800, bs=512):
        if len(buf)<bs: return 0.0
        self.scaler.fit_from_buffer(buf)
        if not self.scaler.fitted: return 0.0
        mu_s  = torch.tensor(self.scaler.mu_s, device=device); std_s = torch.tensor(self.scaler.std_s, device=device)
        mu_ds = torch.tensor(self.scaler.mu_ds,device=device); std_ds= torch.tensor(self.scaler.std_ds,device=device)
        losses=[]
        for _ in range(steps):
            s,a,r,sp,d = buf.sample(bs)
            aoh = F.one_hot(a,num_classes=self.a_dim).float()
            ds = sp - s
            sn  = (s  - mu_s)/std_s
            dsn = (ds - mu_ds)/std_ds
            rn  = (r - self.scaler.mu_r)/self.scaler.std_r
            for m,opt in zip(self.models,self.optims):
                p_ds,p_r,p_d = m(sn,aoh)
                loss = F.mse_loss(p_ds,dsn) + F.mse_loss(p_r,rn) + F.binary_cross_entropy(p_d,d)
                opt.zero_grad(); loss.backward(); torch.nn.utils.clip_grad_norm_(m.parameters(),1.0); opt.step()
                losses.append(loss.item())
        return float(np.mean(losses)) if losses else 0.0
    @torch.no_grad()
    def predict(self,s,a):
        assert self.scaler.fitted
        mu_s  = torch.tensor(self.scaler.mu_s, device=device); std_s = torch.tensor(self.scaler.std_s, device=device)
        mu_ds = torch.tensor(self.scaler.mu_ds,device=device); std_ds= torch.tensor(self.scaler.std_ds,device=device)
        aoh = F.one_hot(a,num_classes=self.a_dim).float()
        sn = (s - mu_s)/std_s
        spL, rL, dL = [],[],[]
        for m in self.models:
            dsn,rn,d = m(sn,aoh)
            ds = dsn*std_ds + mu_ds; r = rn*self.scaler.std_r + self.scaler.mu_r
            spL.append(s+ds); rL.append(r); dL.append(d)
        sp = torch.stack(spL,0); rr=torch.stack(rL,0); dd=torch.stack(dL,0)
        sp_m, rr_m, dd_m = sp.mean(0), rr.mean(0), dd.mean(0)
        uncert = sp.std(0).norm(dim=1) + rr.std(0)
        return sp_m, rr_m, dd_m, uncert

def rollout_model(agent, world, start_states, horizon=1, eps_model=0.05, keep_quantile=0.6):
    B = start_states.size(0); s=start_states.clone(); out=[]
    for _ in range(horizon):
        with torch.no_grad():
            q=agent.q(s); a=q.argmax(1)
            m = torch.rand(B,device=device) < eps_model
            if m.any(): a[m] = torch.randint(0, q.shape[1], (int(m.sum()),), device=device)
        sp,r,d,u = world.predict(s,a)
        thr=torch.quantile(u,keep_quantile); keep = u<=thr
        s_k,a_k,sp_k,r_k,d_k = s[keep],a[keep],sp[keep],r[keep],d[keep]
        s_np, sp_np, r_np, d_np = s_k.cpu().numpy(), sp_k.cpu().numpy(), r_k.cpu().numpy(), d_k.cpu().numpy()
        for i in range(s_np.shape[0]):
            out.append(Transition(clip_obs(s_np[i]), int(a_k[i].item()), float(r_np[i]), clip_obs(sp_np[i]), float(d_np[i]>0.5)))
        s=sp
    return out

# ----------------- Entrenamiento paralelo con VecEnv
def plot_rewards(rs,title=""):
    plt.figure(figsize=(8,4))
    plt.plot(rs,alpha=0.5,label="Return por episodio")
    if len(rs)>=100:
        mov=np.convolve(rs,np.ones(100)/100,mode="valid")
        plt.plot(range(99,99+len(mov)),mov,label="Media móvil (100)")
    plt.title(title); plt.xlabel("Episodio"); plt.ylabel("Return"); plt.grid(True); plt.legend(); plt.show()

def evaluate(agent, env_id=ENV_ID, episodes=10):
    env=gym.make(env_id); scores=[]
    for _ in range(episodes):
        s,_=env.reset(seed=random.randint(0,10_000)); s=clip_obs(s)
        done=False; tot=0.0
        while not done:
            with torch.no_grad():
                a=int(agent.q(torch.tensor(s,dtype=torch.float32,device=device).unsqueeze(0)).argmax(1).item())
            sp,r,t,tr,_=env.step(a); tot+=r; s=clip_obs(sp); done=t or tr
        scores.append(tot)
    env.close(); return float(np.mean(scores)), scores

def train_parallel(
    env_id=ENV_ID,
    num_envs=8,                    # ★ paralelo (ajusta a tus CPUs)
    episodes=1200,                 # episodios totales (sumando los 8)
    use_model=True,                # False = baseline rápido
    warmup_steps=10_000,           # pasos REALES antes de usar WM
    batch_size=256,
    updates_per_step=1,            # SGD por paso vectorizado
    real_batch_frac=0.85,          # mezcla real:modelo
    wm_train_every=20_000,         # entrena WM cada X transiciones
    wm_train_steps=800,
    rollout_every=20_000,          # rollouts cada X transiciones
    rollout_start_states=4096,
    rollout_horizon=1
):
    venv = gym.vector.make(env_id, num_envs=num_envs, asynchronous=True)
    states, _ = venv.reset(seed=SEED)
    states = np.stack([clip_obs(s) for s in states])
    s_dim = states.shape[1]; a_dim = venv.single_action_space.n

    agent = DQNAgent(s_dim, a_dim)
    world = DynEnsemble(E=5)
    real_buf, model_buf = ReplayBuffer(500_000), ReplayBuffer(500_000)

    ep_returns = np.zeros(num_envs, dtype=np.float32)
    finished_returns = []
    total_steps = 0

    pbar = trange(episodes, desc=("DDQN+WM" if use_model else "DDQN baseline"), ncols=120)
    ep_done_count = 0

    while ep_done_count < episodes:
        # ---- actuar en paralelo
        actions = agent.act_batch(states)
        next_states, rewards, terms, truncs, infos = venv.step(actions)
        dones = np.logical_or(terms, truncs)

        # ---- guardar y acumular returns
        for i in range(num_envs):
            real_buf.push(states[i], int(actions[i]), float(rewards[i]), clip_obs(next_states[i]), float(dones[i]))
            ep_returns[i] += rewards[i]
            if dones[i]:
                finished_returns.append(float(ep_returns[i]))
                ep_returns[i] = 0.0
                ep_done_count += 1
                if ep_done_count <= episodes:
                    pbar.update(1)
                    avg100 = float(np.mean(finished_returns[-100:])) if len(finished_returns)>=1 else 0.0
                    pbar.set_postfix(avg100=f"{avg100:7.1f}", eps=f"{agent.eps:.3f}", real=len(real_buf), model=len(model_buf))

        states = np.stack([clip_obs(s) for s in next_states])
        total_steps += num_envs

        # ---- entrenar DQN
        if len(real_buf) >= batch_size and (not use_model or len(model_buf) >= batch_size*(1-real_batch_frac)):
            for _ in range(updates_per_step):
                if use_model and total_steps > warmup_steps:
                    rb = int(batch_size*real_batch_frac); mb = batch_size-rb
                    br = real_buf.sample(rb); bm = model_buf.sample(mb) if mb>0 and len(model_buf)>=mb else real_buf.sample(mb)
                    batch = (torch.cat([br[0],bm[0]],0), torch.cat([br[1],bm[1]],0),
                             torch.cat([br[2],bm[2]],0), torch.cat([br[3],bm[3]],0), torch.cat([br[4],bm[4]],0))
                    agent.train_step(batch)
                else:
                    agent.train_step(real_buf.sample(batch_size))

        # ---- entrenar WM
        if use_model and total_steps > warmup_steps and total_steps % wm_train_every < num_envs:
            _ = world.train_steps(real_buf, steps=wm_train_steps, bs=512)

        # ---- rollouts cortos y filtrados
        if use_model and total_steps > warmup_steps and total_steps % rollout_every < num_envs:
            if world.scaler.fitted and len(real_buf) >= rollout_start_states:
                s0,_,_,_,_ = real_buf.sample(rollout_start_states)
                synth = rollout_model(agent, world, s0, horizon=rollout_horizon, eps_model=0.05, keep_quantile=0.6)
                for tr in synth: model_buf.push(tr.s,tr.a,tr.r,tr.sp,tr.d)

    venv.close()
    return finished_returns, agent, world

# ----------------- CORRIDAS: baseline y con world model
# Baseline rápido (paralelo). Sube num_envs si tu CPU aguanta.
rewards_base, agent_base, _ = train_parallel(
    env_id=ENV_ID, num_envs=8, episodes=1600, use_model=False,
    updates_per_step=1, batch_size=256
)
plot_rewards(rewards_base, f"Baseline Dueling DDQN — {ENV_ID}")
m_base,_ = evaluate(agent_base, ENV_ID, episodes=10)
print(f"[BASELINE] Return promedio (10 eps): {m_base:.1f}")

# World Model conservador (misma recogida paralela)
rewards_wm, agent_wm, world = train_parallel(
    env_id=ENV_ID, num_envs=8, episodes=2000, use_model=True,
    warmup_steps=12_000, real_batch_frac=0.9,
    wm_train_every=25_000, wm_train_steps=800,
    rollout_every=25_000, rollout_start_states=4096,
    rollout_horizon=1, updates_per_step=1, batch_size=256
)
plot_rewards(rewards_wm, f"Dueling DDQN + World Model — {ENV_ID}")
m_wm,_ = evaluate(agent_wm, ENV_ID, episodes=10)
print(f"[WORLD MODEL] Return promedio (10 eps): {m_wm:.1f}")
# ===================================================================================================
```
